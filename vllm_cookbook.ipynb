## Repeat a span of layers during execution

The function below runs **batched** autoregressive generation while repeating decoder layers in the inclusive range 

- 
	i = start layer index
- 
	num_repeat_layers = number of layers to repeat (so end index is j = i + num_repeat_layers - 1)

Behavior: for each forward step, the model executes layers `0..j`, then loops back and re-runs layers `i..j` for `extra_passes` additional times, then continues with layers `j+1..end` as normal.

This is implemented for **decoder-only Transformer stacks** that expose `model.model.embed_tokens`, `model.model.layers`, and `model.model.norm` (Llama-like). If your architecture differs, the small `get_stack(model)` helper is the place to adapt.

````python
import inspect
from dataclasses import dataclass
from typing import Callable, List, Optional, Sequence, Tuple

import torch


@dataclass
class _DecoderStack:
    embed_tokens: torch.nn.Module
    layers: torch.nn.ModuleList
    norm: torch.nn.Module
    lm_head: torch.nn.Module


def _first_attr(obj, names: Sequence[str]) -> Optional[torch.nn.Module]:
    for n in names:
        if hasattr(obj, n):
            return getattr(obj, n)
    return None


def _as_modulelist(x) -> torch.nn.ModuleList:
    # Some models use plain list; normalize for indexing/len
    if isinstance(x, torch.nn.ModuleList):
        return x
    if isinstance(x, (list, tuple)):
        return torch.nn.ModuleList(list(x))
    raise ValueError(f"Expected ModuleList/list/tuple for layers, got {type(x)}")


def _get_decoder_stack(model: torch.nn.Module) -> _DecoderStack:
    """Best-effort accessor for decoder stacks.

    Lowest safe repeat unit is the top-level block list used by the model forward loop.
    For Nemotron-H, this is typically a list of hybrid blocks / "superlayers".
    """
    if hasattr(model, "model"):
        inner = model.model

        embed = _first_attr(inner, ("embed_tokens", "tok_embeddings", "wte"))
        norm = _first_attr(inner, ("norm", "ln_f", "final_layernorm", "final_norm"))

        # Prefer explicit superlayer container if present; otherwise fall back to layers.
        layers_obj = None
        for ln in ("superlayers", "layers", "blocks", "h"):
            if hasattr(inner, ln):
                layers_obj = getattr(inner, ln)
                break

        if embed is not None and norm is not None and layers_obj is not None and hasattr(model, "lm_head"):
            return _DecoderStack(
                embed_tokens=embed,
                layers=_as_modulelist(layers_obj),
                norm=norm,
                lm_head=model.lm_head,
            )

    # Common GPT-2 style fallback
    if hasattr(model, "transformer"):
        inner = model.transformer
        embed = _first_attr(inner, ("wte", "embed_tokens"))
        norm = _first_attr(inner, ("ln_f", "norm"))
        layers_obj = getattr(inner, "h", None)
        if embed is not None and norm is not None and layers_obj is not None and hasattr(model, "lm_head"):
            return _DecoderStack(
                embed_tokens=embed,
                layers=_as_modulelist(layers_obj),
                norm=norm,
                lm_head=model.lm_head,
            )

    raise ValueError(
        "Could not locate a known decoder stack. "
        "Inspect your loaded model and adapt _get_decoder_stack() accordingly."
    )


def _call_layer(layer: torch.nn.Module, hidden_states: torch.Tensor, **kwargs):
    """Call a layer with only the kwargs it supports."""
    sig = inspect.signature(layer.forward)
    filtered = {k: v for k, v in kwargs.items() if k in sig.parameters}
    return layer(hidden_states, **filtered)


def forward_repeating_layers(
    model: torch.nn.Module,
    input_ids: torch.Tensor,
    *,
    attention_mask: Optional[torch.Tensor] = None,
    position_ids: Optional[torch.Tensor] = None,
    repeat_span: Tuple[int, int],
    extra_passes: int = 1,
) -> torch.Tensor:
    """Forward pass that repeats layers in repeat_span.

    Returns logits for all positions.

    Notes:
    - This is a **no-cache** forward (no KV cache). It’s intended for small experiments.
    - For large models/long contexts, you’ll want a cache-aware implementation.
    """
    i, j = repeat_span
    if extra_passes < 0:
        raise ValueError("extra_passes must be >= 0")

    stack = _get_decoder_stack(model)
    layers = stack.layers
    n_layers = len(layers)

    if not (0 <= i <= j < n_layers):
        raise ValueError(f"repeat_span must satisfy 0 <= i <= j < {n_layers}. Got {(i, j)}")

    if attention_mask is None:
        attention_mask = torch.ones_like(input_ids, dtype=torch.long, device=input_ids.device)

    # Build a basic position_ids if none given.
    if position_ids is None:
        position_ids = (attention_mask.cumsum(-1) - 1).clamp(min=0)

    hidden_states = stack.embed_tokens(input_ids)

    # Some architectures (including Nemotron-H in Transformers) may use `cache_position`
    # instead of / in addition to `position_ids`.
    cache_position = torch.arange(input_ids.size(1), device=input_ids.device)

    layer_kwargs = {
        "attention_mask": attention_mask,
        "position_ids": position_ids,
        "cache_position": cache_position,
        "use_cache": False,
        "output_attentions": False,
    }

    # 0..j
    for idx in range(0, j + 1):
        out = _call_layer(layers[idx], hidden_states, **layer_kwargs)
        hidden_states = out[0] if isinstance(out, (tuple, list)) else out

    # loop back i..j
    for _ in range(extra_passes):
        for idx in range(i, j + 1):
            out = _call_layer(layers[idx], hidden_states, **layer_kwargs)
            hidden_states = out[0] if isinstance(out, (tuple, list)) else out

    # j+1..end
    for idx in range(j + 1, n_layers):
        out = _call_layer(layers[idx], hidden_states, **layer_kwargs)
        hidden_states = out[0] if isinstance(out, (tuple, list)) else out

    hidden_states = stack.norm(hidden_states)
    logits = stack.lm_head(hidden_states)
    return logits


def generate_batched_with_repeated_layers(
    model: torch.nn.Module,
    tokenizer,
    prompts: Sequence[str],
    *,
    i: int,
    j: int,
    extra_passes: int = 1,
    max_new_tokens: int = 64,
    temperature: float = 0.0,
) -> List[str]:
    """Simple batched generation loop using forward_repeating_layers (no KV cache)."""
    device = next(model.parameters()).device
    enc = tokenizer(
        list(prompts),
        return_tensors="pt",
        padding=True,
        truncation=True,
    )
    input_ids = enc["input_ids"].to(device)
    attention_mask = enc.get("attention_mask", None)
    if attention_mask is not None:
        attention_mask = attention_mask.to(device)

    model.eval()
    with torch.no_grad():
        for _ in range(max_new_tokens):
            logits = forward_repeating_layers(
                model,
                input_ids,
                attention_mask=attention_mask,
                repeat_span=(i, j),
                extra_passes=extra_passes,
            )
            next_logits = logits[:, -1, :]
            if temperature and temperature > 0:
                probs = torch.softmax(next_logits / temperature, dim=-1)
                next_ids = torch.multinomial(probs, num_samples=1)
            else:
                next_ids = torch.argmax(next_logits, dim=-1, keepdim=True)

            input_ids = torch.cat([input_ids, next_ids], dim=-1)
            if attention_mask is not None:
                attention_mask = torch.cat(
                    [attention_mask, torch.ones((attention_mask.size(0), 1), device=device, dtype=attention_mask.dtype)],
                    dim=-1,
                )

    return tokenizer.batch_decode(input_ids, skip_special_tokens=True)
````

### Benchmark: repeated layers vs control (quality-only)

This section benchmarks `generate_batched_with_repeated_layers()` against a **control** (same model, same prompts, no repeated layers) using:

- **Next-token KL divergence** between variant and control distributions (lower = closer to control)
- Short continuation **NLL** (negative log-likelihood) on a fixed reference continuation (lower = better on that continuation)

Tip: use a smaller model (e.g. `gpt2` or `meta-llama/Llama-3.2-1B` if you have access) for fast iteration.

````python
from typing import Dict

import torch
import torch.nn.functional as F


def forward_control_no_repeat(
    model: torch.nn.Module,
    input_ids: torch.Tensor,
    *,
    attention_mask: torch.Tensor | None = None,
    position_ids: torch.Tensor | None = None,
) -> torch.Tensor:
    """Plain forward through all decoder layers once (no KV cache)."""
    stack = _get_decoder_stack(model)
    layers = stack.layers

    if attention_mask is None:
        attention_mask = torch.ones_like(input_ids, dtype=torch.long, device=input_ids.device)
    if position_ids is None:
        position_ids = (attention_mask.cumsum(-1) - 1).clamp(min=0)

    hidden_states = stack.embed_tokens(input_ids)
    layer_kwargs = {
        "attention_mask": attention_mask,
        "position_ids": position_ids,
        "use_cache": False,
        "output_attentions": False,
    }

    for layer in layers:
        out = _call_layer(layer, hidden_states, **layer_kwargs)
        hidden_states = out[0] if isinstance(out, (tuple, list)) else out

    hidden_states = stack.norm(hidden_states)
    return stack.lm_head(hidden_states)


def _next_token_kl(logits_p: torch.Tensor, logits_q: torch.Tensor) -> torch.Tensor:
    """KL(p || q) for the last position, per batch element."""
    p = F.log_softmax(logits_p[:, -1, :], dim=-1)
    q = F.log_softmax(logits_q[:, -1, :], dim=-1)
    p_prob = p.exp()
    return (p_prob * (p - q)).sum(dim=-1)


def continuation_nll(
    logits: torch.Tensor,
    input_ids: torch.Tensor,
    *,
    prompt_lens: torch.Tensor,
) -> torch.Tensor:
    """Average NLL over the continuation tokens (after each prompt length)."""
    shift_logits = logits[:, :-1, :].contiguous()
    shift_labels = input_ids[:, 1:].contiguous()

    seq_len_minus1 = shift_labels.size(1)
    arange = torch.arange(seq_len_minus1, device=input_ids.device).unsqueeze(0)
    cont_mask = arange >= (prompt_lens.unsqueeze(1) - 1).clamp(min=0)

    per_token_nll = F.cross_entropy(
        shift_logits.view(-1, shift_logits.size(-1)),
        shift_labels.view(-1),
        reduction="none",
    ).view(shift_labels.size())

    masked = per_token_nll * cont_mask
    denom = cont_mask.sum(dim=1).clamp(min=1)
    return masked.sum(dim=1) / denom


def benchmark_repeated_layers_quality(
    *,
    model_name: str = "gpt2",
    device: str | None = None,
    prompts: list[str] | None = None,
    reference_continuation: str = "\nThe answer is:",
    i: int = 2,
    j: int = 4,
    extra_passes: int = 2,
) -> Dict[str, float]:
    """Quality-only probes: next-token KL + continuation NLL (variant vs control)."""
    from transformers import AutoModelForCausalLM, AutoTokenizer

    if prompts is None:
        prompts = [
            "Explain in one sentence what a KV cache is.",
            "The capital of France is",
            "Write a short tagline for a GPU cloud:",
            "2+2=",
        ]

    tok = AutoTokenizer.from_pretrained(model_name)
    if tok.pad_token_id is None:
        tok.pad_token = tok.eos_token

    if device is None:
        device = "cuda" if torch.cuda.is_available() else "cpu"

    dtype = (
        torch.bfloat16
        if device == "cuda" and torch.cuda.is_bf16_supported()
        else torch.float16
        if device == "cuda"
        else torch.float32
    )
    model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=dtype).to(device)

    # Build inputs: prompt + fixed continuation
    full_texts = [p + reference_continuation for p in prompts]
    enc_full = tok(full_texts, return_tensors="pt", padding=True)
    enc_prompt = tok(prompts, return_tensors="pt", padding=True)

    input_ids = enc_full["input_ids"].to(device)
    attention_mask = enc_full["attention_mask"].to(device)
    prompt_lens = enc_prompt["attention_mask"].sum(dim=1).to(device)

    model.eval()
    with torch.no_grad():
        logits_control = forward_control_no_repeat(model, input_ids, attention_mask=attention_mask)
        logits_repeat = forward_repeating_layers(
            model,
            input_ids,
            attention_mask=attention_mask,
            repeat_span=(i, j),
            extra_passes=extra_passes,
        )

        kl = _next_token_kl(logits_repeat, logits_control)
        nll_control = continuation_nll(logits_control, input_ids, prompt_lens=prompt_lens)
        nll_repeat = continuation_nll(logits_repeat, input_ids, prompt_lens=prompt_lens)

    return {
        "next_token_kl_mean": kl.mean().item(),
        "next_token_kl_median": kl.median().item(),
        "nll_control_mean": nll_control.mean().item(),
        "nll_repeat_mean": nll_repeat.mean().item(),
        "nll_delta_mean": (nll_repeat - nll_control).mean().item(),
        "batch_size": float(len(prompts)),
    }


# Example run (small model by default)
results = benchmark_repeated_layers_quality(model_name="gpt2", i=2, j=4, extra_passes=2)
for k, v in results.items():
    print(f"{k}: {v}")
````

### Objective MCQ benchmarks (GPQA / ARC-Challenge / HellaSwag)

This evaluation computes **multiple-choice accuracy** by selecting the option with the **highest conditional log-likelihood** (equivalently: lowest mean NLL over the option tokens).

It reports **separate scores** for:
- **GPQA** (often gated; will be skipped with an actionable message if unavailable)
- **ARC-Challenge (ARC-2)**
- **HellaSwag**

It also reports each benchmark’s **baseline vs repeated-layer** accuracy and the delta.

````python
import math
import random
from dataclasses import dataclass
from typing import Callable, Iterable, List, Optional, Sequence, Tuple

import torch
import torch.nn.functional as F


@dataclass
class MCQExample:
    benchmark: str
    question: str
    choices: List[str]
    correct_index: int


def _try_import_datasets():
    try:
        import datasets  # type: ignore

        return datasets
    except Exception as e:  # noqa: BLE001
        raise RuntimeError(
            "Missing dependency: `datasets`. Install with `%pip install -U datasets`."
        ) from e


def _sample_indices(n: int, k: int, seed: int) -> List[int]:
    k = min(k, n)
    rng = random.Random(seed)
    return rng.sample(range(n), k)


def load_arc_challenge(max_examples: int = 200, seed: int = 0) -> List[MCQExample]:
    datasets = _try_import_datasets()
    ds = datasets.load_dataset("allenai/ai2_arc", "ARC-Challenge", split="validation")
    idxs = _sample_indices(len(ds), max_examples, seed)

    out: List[MCQExample] = []
    for ix in idxs:
        ex = ds[int(ix)]
        q = ex["question"]
        labels = ex["choices"]["label"]
        texts = ex["choices"]["text"]
        answer_key = ex["answerKey"]
        correct_index = labels.index(answer_key)
        out.append(MCQExample("arc_challenge", q, list(texts), int(correct_index)))
    return out


def load_hellaswag(max_examples: int = 200, seed: int = 0) -> List[MCQExample]:
    datasets = _try_import_datasets()
    ds = datasets.load_dataset("Rowan/hellaswag", split="validation")
    idxs = _sample_indices(len(ds), max_examples, seed)

    out: List[MCQExample] = []
    for ix in idxs:
        ex = ds[int(ix)]
        # Commonly used fields
        ctx = ex.get("ctx", "")
        # Build a readable question/context
        q = ctx.strip() if ctx else (ex.get("ctx_a", "") + " " + ex.get("ctx_b", "")).strip()
        endings = list(ex["endings"])
        correct_index = int(ex["label"])
        out.append(MCQExample("hellaswag", q, endings, correct_index))
    return out


def _infer_gpqa_fields(record: dict) -> Optional[Tuple[str, List[str], int]]:
    """Best-effort schema adapter for GPQA variants.

    Returns (question, choices, correct_index) or None if unrecognized.
    """
    # Common formatted schema (some community re-formats)
    if "question" in record and "choices" in record and "answer" in record:
        question = record["question"]
        choices = list(record["choices"])
        ans = record["answer"]
        if isinstance(ans, int):
            return question, choices, int(ans)
        if isinstance(ans, str) and ans in choices:
            return question, choices, choices.index(ans)

    # Original CSV-like schema: correct + three incorrects
    # (names vary; we try a few)
    q_key = "question" if "question" in record else "Question" if "Question" in record else None
    if q_key is None:
        return None

    # Find correct answer key
    for ck in ("correct_answer", "Correct Answer", "correct", "answer", "Answer"):
        if ck in record:
            correct = record[ck]
            break
    else:
        correct = None

    if correct is None:
        return None

    incorrects: List[str] = []
    # Try lists
    for ik in ("incorrect_answers", "Incorrect Answers", "incorrect", "distractors"):
        if ik in record and isinstance(record[ik], (list, tuple)):
            incorrects = list(record[ik])
            break

    # Try enumerated columns
    if not incorrects:
        for ik in (
            "incorrect_answer_1",
            "incorrect_answer_2",
            "incorrect_answer_3",
            "Incorrect Answer 1",
            "Incorrect Answer 2",
            "Incorrect Answer 3",
        ):
            if ik in record:
                incorrects.append(record[ik])

    if len(incorrects) < 3:
        return None

    choices = [correct] + incorrects[:3]
    correct_index = 0
    return str(record[q_key]), [str(c) for c in choices], int(correct_index)


def load_gpqa(
    config: str = "gpqa_main",
    max_examples: int = 200,
    seed: int = 0,
) -> Tuple[List[MCQExample], Optional[str]]:
    """Loads GPQA if accessible; otherwise returns ([], error_message)."""
    datasets = _try_import_datasets()
    try:
        ds = datasets.load_dataset("Idavidrein/gpqa", config, split="train")
    except Exception as e:  # noqa: BLE001
        return [], (
            "GPQA could not be loaded (it may be gated). "
            "If you have access, run `huggingface-cli login` and accept the dataset terms on Hugging Face. "
            f"Original error: {type(e).__name__}: {e}"
        )

    idxs = _sample_indices(len(ds), max_examples, seed)
    out: List[MCQExample] = []
    for ix in idxs:
        rec = dict(ds[int(ix)])
        parsed = _infer_gpqa_fields(rec)
        if parsed is None:
            continue
        q, choices, correct_index = parsed
        out.append(MCQExample("gpqa", q, choices, correct_index))

    if not out:
        return [], "GPQA loaded but schema was not recognized by this notebook. We can adapt the parser once we see its columns."

    return out, None


def build_mcq_prompt(question: str, choices: Sequence[str]) -> str:
    labels = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lines = ["Choose the best answer.", "", f"Question: {question}", "", "Options:"]
    for i, c in enumerate(choices):
        lines.append(f"{labels[i]}. {c}")
    lines.append("")
    lines.append("Answer:")
    return "\n".join(lines)
````

````python
def _prepare_mcq_batch(
    tokenizer,
    prompts: Sequence[str],
    continuations: Sequence[str],
    device: torch.device,
):
    """Tokenize prompt+continuation and build labels masking prompt tokens."""
    assert len(prompts) == len(continuations)

    prompt_enc = tokenizer(list(prompts), return_tensors="pt", padding=True, truncation=True)
    full_texts = [p + c for p, c in zip(prompts, continuations)]
    full_enc = tokenizer(full_texts, return_tensors="pt", padding=True, truncation=True)

    input_ids = full_enc["input_ids"].to(device)
    attention_mask = full_enc.get("attention_mask")
    if attention_mask is not None:
        attention_mask = attention_mask.to(device)

    prompt_lens = prompt_enc.get("attention_mask")
    if prompt_lens is None:
        # Fall back to counting non-pad tokens in prompt input_ids
        pad_id = tokenizer.pad_token_id
        prompt_lens = (prompt_enc["input_ids"] != pad_id).long()
    prompt_lens = prompt_lens.sum(dim=1).to(device)

    # Labels: ignore everything up through the prompt
    labels = input_ids.clone()
    labels[:, :] = -100

    # We want to score the continuation tokens only.
    # Note: labels correspond to target tokens; the model predicts token t using logit at t-1.
    # A simple and effective way: unmask tokens at indices >= prompt_len.
    for b in range(labels.size(0)):
        pl = int(prompt_lens[b].item())
        labels[b, pl:] = input_ids[b, pl:]

    return input_ids, attention_mask, labels


def mean_nll_from_logits(logits: torch.Tensor, labels: torch.Tensor) -> torch.Tensor:
    """Mean NLL per batch element over non-ignored label positions."""
    shift_logits = logits[:, :-1, :].contiguous()
    shift_labels = labels[:, 1:].contiguous()

    vocab = shift_logits.size(-1)
    loss_flat = F.cross_entropy(
        shift_logits.view(-1, vocab),
        shift_labels.view(-1),
        reduction="none",
        ignore_index=-100,
    ).view(shift_labels.size())

    mask = shift_labels != -100
    denom = mask.sum(dim=1).clamp(min=1)
    return (loss_flat * mask).sum(dim=1) / denom


def score_mcq_examples(
    model: torch.nn.Module,
    tokenizer,
    examples: Sequence[MCQExample],
    *,
    forward_fn: Callable[..., torch.Tensor],
    batch_size: int = 8,
    device: str | None = None,
    repeat_span: Tuple[int, int] | None = None,
    extra_passes: int = 0,
) -> float:
    """Returns accuracy on MCQ examples using log-likelihood selection.

    If repeat_span is provided, it is passed to forward_repeating_layers.
    """
    if device is None:
        device = "cuda" if torch.cuda.is_available() else "cpu"
    dev = torch.device(device)

    model.eval()

    correct = 0
    total = 0

    # Flatten (example, choice) into candidate rows
    for start in range(0, len(examples), batch_size):
        batch = examples[start : start + batch_size]

        cand_prompts: List[str] = []
        cand_conts: List[str] = []
        cand_choice_counts: List[int] = []
        correct_indices: List[int] = []

        for ex in batch:
            prompt = build_mcq_prompt(ex.question, ex.choices)
            # Continuations are just the answer letter (A/B/...) + newline
            labels = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
            conts = [f" {labels[k]}\n" for k in range(len(ex.choices))]

            cand_prompts.extend([prompt] * len(conts))
            cand_conts.extend(conts)
            cand_choice_counts.append(len(conts))
            correct_indices.append(ex.correct_index)

        input_ids, attention_mask, labels_tensor = _prepare_mcq_batch(
            tokenizer, cand_prompts, cand_conts, dev
        )

        with torch.no_grad():
            if repeat_span is None:
                logits = forward_fn(model, input_ids, attention_mask=attention_mask)
            else:
                logits = forward_fn(
                    model,
                    input_ids,
                    attention_mask=attention_mask,
                    repeat_span=repeat_span,
                    extra_passes=extra_passes,
                )

            nll = mean_nll_from_logits(logits, labels_tensor)

        # Unflatten
        offset = 0
        for ex_i, n_choices in enumerate(cand_choice_counts):
            scores = nll[offset : offset + n_choices]
            pred = int(torch.argmin(scores).item())
            if pred == int(correct_indices[ex_i]):
                correct += 1
            total += 1
            offset += n_choices

    return correct / max(total, 1)
````

````python
def run_objective_mcq_benchmarks(
    *,
    model_name: str = "gpt2",
    device: str | None = None,
    i: int = 2,
    j: int = 4,
    extra_passes: int = 2,
    max_examples_per_dataset: int = 200,
    seed: int = 0,
    gpqa_config: str = "gpqa_main",
    batch_size: int = 8,
):
    """Print separate MCQ accuracies for GPQA, ARC-Challenge, HellaSwag.

    Uses log-likelihood selection (no generation), comparing control vs repeated layers.
    """
    from transformers import AutoModelForCausalLM, AutoTokenizer

    if device is None:
        device = "cuda" if torch.cuda.is_available() else "cpu"

    tok = AutoTokenizer.from_pretrained(model_name)
    if tok.pad_token_id is None:
        tok.pad_token = tok.eos_token

    dtype = (
        torch.bfloat16
        if device == "cuda" and torch.cuda.is_bf16_supported()
        else torch.float16
        if device == "cuda"
        else torch.float32
    )
    model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=dtype).to(device)

    # Load datasets (subset for speed)
    arc = load_arc_challenge(max_examples=max_examples_per_dataset, seed=seed)
    hs = load_hellaswag(max_examples=max_examples_per_dataset, seed=seed)
    gpqa, gpqa_err = load_gpqa(config=gpqa_config, max_examples=max_examples_per_dataset, seed=seed)

    repeat_span = (i, j)

    def eval_one(name: str, examples: List[MCQExample]):
        acc_control = score_mcq_examples(
            model,
            tok,
            examples,
            forward_fn=forward_control_no_repeat,
            batch_size=batch_size,
            device=device,
        )
        acc_repeat = score_mcq_examples(
            model,
            tok,
            examples,
            forward_fn=forward_repeating_layers,
            repeat_span=repeat_span,
            extra_passes=extra_passes,
            batch_size=batch_size,
            device=device,
        )
        delta = acc_repeat - acc_control
        print(
            f"{name} accuracy: control={acc_control:.4f} | repeat(i={i},j={j},passes={extra_passes})={acc_repeat:.4f} | delta={delta:+.4f}"
        )

    print("\nObjective MCQ benchmarks (log-likelihood selection)")
    print(f"Model: {model_name} | device={device}")
    print(f"Subset size per dataset: {max_examples_per_dataset} | seed={seed} | batch_size={batch_size}")
    print()

    eval_one("ARC-Challenge", arc)
    eval_one("HellaSwag", hs)

    if gpqa:
        eval_one(f"GPQA ({gpqa_config})", gpqa)
    else:
        print(f"GPQA ({gpqa_config}): skipped. {gpqa_err}")


# Example run (small model by default)
run_objective_mcq_benchmarks(
    model_name="gpt2",
    i=2,
    j=4,
    extra_passes=2,
    max_examples_per_dataset=50,
    seed=0,
    batch_size=8,
)
````

### Controlling Reasoning Budget

The `reasoning_budget` parameter allows you to limit the length of the model's reasoning trace. When the reasoning output reaches the specified token budget, the model will attempt to gracefully end the reasoning at the next newline character. 

If no newline is encountered within 500 tokens after reaching the budget threshold, the reasoning trace will be forcibly terminated at `reasoning_budget + 500` tokens to prevent excessive generation.


````python
from typing import Any, Dict, List
import openai
from transformers import AutoTokenizer


class ThinkingBudgetClient:
    def __init__(self, base_url: str, api_key: str, tokenizer_name_or_path: str):
        self.base_url = base_url
        self.api_key = api_key
        self.tokenizer = AutoTokenizer.from_pretrained(tokenizer_name_or_path)
        self.client = openai.OpenAI(base_url=self.base_url, api_key=self.api_key)

    def chat_completion(
        self,
        model: str,
        messages: List[Dict[str, Any]],
        reasoning_budget: int = 512,
        max_tokens: int = 1024,
        **kwargs,
    ) -> Dict[str, Any]:
        assert (
            max_tokens > reasoning_budget
        ), f"reasoning_budget must be smaller than max_tokens. Given {max_tokens=} and {reasoning_budget=}"

        # 1. first call chat completion to get reasoning content
        response = self.client.chat.completions.create(
            model=model, 
            messages=messages, 
            max_tokens=reasoning_budget, 
            **kwargs
        )
    
        reasoning_content = response.choices[0].message.reasoning_content or ""
        
        if "</think>" not in reasoning_content:
            # reasoning content is too long, closed with a period (.)
            reasoning_content = f"{reasoning_content}.\n</think>\n\n"
        
        reasoning_tokens_used = len(
            self.tokenizer.encode(reasoning_content, add_special_tokens=False)
        )
        remaining_tokens = max_tokens - reasoning_tokens_used
        
        assert (
            remaining_tokens > 0
        ), f"remaining tokens must be positive. Given {remaining_tokens=}. Increase max_tokens or lower reasoning_budget."

        # 2. append reasoning content to messages and call completion
        messages.append({"role": "assistant", "content": reasoning_content})
        prompt = self.tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            continue_final_message=True,
        )
        
        response = self.client.completions.create(
            model=model, 
            prompt=prompt, 
            max_tokens=remaining_tokens, 
            **kwargs
        )

        response_data = {
            "reasoning_content": reasoning_content.strip().strip("</think>").strip(),
            "content": response.choices[0].text,
            "finish_reason": response.choices[0].finish_reason,
        }
        return response_data
````
